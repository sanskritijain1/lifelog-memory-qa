```python
import pandas as pd

df = pd.read_csv("epic-kitchens-100-annotations/EPIC_100_train.csv")
df.head()
```

```python
video_stats = df.groupby("video_id").agg(
    num_actions=("verb", "count"),
    unique_actions=("verb", "nunique"),
    unique_objects=("noun", "nunique")
).reset_index()
```

```python
video_stats["score"] = (
    video_stats["num_actions"] * 0.5 +
    video_stats["unique_actions"] * 0.3 +
    video_stats["unique_objects"] * 0.2
)

top_videos = video_stats.sort_values("score", ascending=False).head(5)
selected_videos = top_videos["video_id"].tolist()

selected_videos
```

```python
with open("selected_videos.txt", "w") as f:
    for v in selected_videos:
        f.write(v + "\n")
```
